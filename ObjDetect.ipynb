```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Upload the picture inside which you want to find the object!
full=cv2.imread('DATA/image.png')
full=cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
#Upload the exact picture of the object you want to find!
face=cv2.imread('DATA/mouse.png')
face=cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    #Create a copy 
    full_copy=full.copy()

    method=eval(m)
    #Template Matching
    res=cv2.matchTemplate(full_copy,face,method)

    min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)

    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=min_loc
    else:
        top_left=max_loc

    height,width,channels=face.shape
    bottom_right=(top_left[0]+width,top_left[1]+height)

    cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)
    plt.subplot(121)
    plt.imshow(res)
    plt.title("HEATMAP OF TEMPLATE MATCHING")
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('DETECTION OF TEMPLATE')
    #TITLE WHICH THE METHOD USED
    plt.suptitle(m)
    plt.show()
```

```python
my_method=eval('cv2.TM_CCOEFF')
res=cv2.matchTemplate(full,face,my_method)
```
